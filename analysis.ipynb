# Iris Data Analysis
---
This notebook contains details and analysis on the popular Iris data set.

## Background
---

The iris data set was collected by the botanist Edgar Anderson in 1935 and when it first appeared in publications, it was used to illustrate what was at the time, a new technique of linear discriminant analysis [^1]. The data contains the measurements of the lengths and widths of the sepals and petals of three different species of iris flowers; virginica, versicolor, and setosa [^1]. Fifty plants wre measured of each species giving 150 pieces of data in the set [^1].

[^1]: https://doi.org/10.1111/1740-9713.01589

## The Data
---

```python
#first step is to import pandas
import pandas as pd
```

```python
#I will then import the csv file containing the raw data
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
#Next step is to open the data
df
```

```python
#Use df.describe() to summarise the data and bring up the mean, median, standard deviation, etc.
df.describe()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style 
import seaborn as sns
```

```python
# For 3(1) and 3(2): use the write open() function and use modes "a" for appending or "w" for writing. 
# Append with add to the end of an existing file and write will overwrite any existing file.
# Both of these creates a file if the file doesn't exist.
# For text files use "t" and for binary files such as images use "b".
```

```python
# scatterplot using seaborn
sns.scatterplot(x='sepal_length', y='sepal_width', 
				hue='species', data=df, ) 
# Placing Legend outside the Figure 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between sepal width and sepal length')
plt.show()


sns.scatterplot(x='petal_length', y='petal_width', 
				hue='species', data=df, ) 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between petal width and petal length')
plt.show()


sns.scatterplot(x='petal_length', y='sepal_length', 
				hue='species', data=df, ) 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between sepal length and petal length')
plt.show()


sns.scatterplot(x='petal_length', y='sepal_width', 
				hue='species', data=df, ) 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between sepal width and petal length')
plt.show()


sns.scatterplot(x='petal_width', y='sepal_width', 
				hue='species', data=df, ) 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between petal width and sepal width')
plt.show()


sns.scatterplot(x='petal_width', y='sepal_length', 
				hue='species', data=df, ) 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.title('Correlation between petal width and sepal length')
plt.show()

# This method is tedious so I will use a quicker way of displaying all the scatterplots at once.
```

```python
# Pairplot is used for multivariate analysis
# Is used to visualise the relationship between all the variables in the data
g = sns.pairplot(df,hue="species") 
```

```python
g = sns.boxplot(y='species', x='sepal_length', data=df)
plt.title('Boxplot of Sepal Length Distribution')
plt.show()

g = sns.boxplot(y='species', x='sepal_width', data=df)
plt.title('Boxplot of Sepal Width Distribution')
plt.show()

g = sns.boxplot(y='species', x='petal_length', data=df)
plt.title('Boxplot of Petal Length Distribution')
plt.show()

g = sns.boxplot(y='species', x='petal_width', data=df)
plt.title('Boxplot of Petal Width Distribution')
plt.show()
```
